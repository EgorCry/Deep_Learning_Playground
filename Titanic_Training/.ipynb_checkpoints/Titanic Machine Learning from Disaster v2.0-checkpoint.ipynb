```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_file = pd.read_csv('Datasets/train.csv')
```

```python
train_file.head(10)
```

```python
train_file = train_file.drop('Embarked', axis=1)
train_file['FamilySize'] = train_file['SibSp'] + train_file['Parch'] + 1
train_file = train_file.drop('SibSp', axis=1)
train_file = train_file.drop('Parch', axis=1)
```

```python
train_file.corr()
```

```python
train_file['Age'].isnull().values.any()
```

```python
ticket_typeList = np.unique(train_file['Ticket'].apply(lambda x: x.split()[0] if not x.split()[0].isdigit() else 'None'))
ticket_typeList = np.append( ticket_typeList[:17], ticket_typeList[18:])
dict_ticket_typeList = {}
for i in range(len(ticket_typeList)):
    dict_ticket_typeList.update({ticket_typeList[i]: i+1})
print(dict_ticket_typeList)
```

```python
train_file.describe()
```

```python
train_file.info()
```

```python
train_file["Ticket_Type"] = np.where(train_file["Ticket"].str.isdigit(), 0, train_file["Ticket"])
train_file['Ticket_Type'] = train_file['Ticket'].apply(lambda x: dict_ticket_typeList[x.split()[0]] if 
                                                      x.split()[0] in dict_ticket_typeList.keys() else 0)
train_file['Sex'] = train_file['Sex'].map({'female': 1, 'male': 0})
train_file['Age'].fillna(value=float(train_file['Age'].median()), inplace=True)
train_file.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_list = ['Pclass', 'Age', 'Fare', 'FamilySize', 'Ticket_Type', 'Sex']
x_train = tf.convert_to_tensor(scaler.fit_transform(train_file[train_list].astype('float64')))

y_train = train_file['Survived']

print(x_train[0])
x_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
def build_model():
    model = Sequential(name='Titanic')
    model.add(Dense(2048, activation='relu', input_dim=(6)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(2048, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=10, validation_split=0.2, verbose=2)
```
