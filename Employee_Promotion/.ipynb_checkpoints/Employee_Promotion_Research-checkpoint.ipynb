```python
ed = pd.get_dummies(data.education)
ed
```

```python
data = pd.concat([data, ed], axis=1)
data
```

```python
data.boxplot(column=["education"])
```

```python
data = data.drop(columns=['education'])
data
```

```python
d_ed = {}
i = 0
for j in data.education.unique():
    d_ed[j] = i
    i += 1
d_ed
```

```python
data.education = data.education.replace(d_ed)
data
```

```python
d_rc = {}
i = 0
for j in data.recruitment_channel.unique():
    d_rc[j] = i
    i += 1
d_rc
```

```python
data.recruitment_channel = data.recruitment_channel.replace(d_rc)
data
```

```python
d_reg = {}
i = 0
for j in data.region.unique():
    d_reg[j] = i
    i += 1
d_reg
```

```python
data.region = data.region.replace(d_reg)
data
```

```python
d_dep = {}
i = 0
for j in data.department.unique():
    d_dep[j] = i
    i += 1
d_dep
```

```python
data.department = data.department.replace(d_dep)
data
```

```python
d_gen = {}
i = 0
for j in data.gender.unique():
    d_gen[j] = i
    i += 1
d_gen
```

```python
data.gender = data.gender.replace(d_gen)
data
```

```python
data.isna().sum()
```

```python
data.previous_year_rating = data.previous_year_rating.fillna(0.0)
data.avg_training_score = data.avg_training_score.fillna(data.avg_training_score.mean())
data
```

```python
data = data.drop(columns=['department', 'region', 'previous_year_rating', 'avg_training_score'])
data
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
k = len(data.columns)
n = 2
m = (k - 1) // n + 1
fig, axes = plt.subplots(m, n, figsize=(n * 5, m * 3))
for i, (name, col) in enumerate(data.iteritems()):
    r, c = i // n, i % n
    ax = axes[r, c]
    col.hist(ax=ax)
    ax2 = col.plot.kde(ax=ax, secondary_y=True, title=name)
    ax2.set_ylim(0)

fig.tight_layout()
```

```python
print(data[data.gender == 1].shape)
print(data[(data.gender == 1) & (data.is_promoted == 1)].shape)
per_men = round(len(data[(data.gender == 1) & (data.is_promoted == 1)]) / len(data[data.gender == 1]) * 100, 2)
print(data[data.gender == 0].shape) 
print(data[(data.gender == 0) & (data.is_promoted == 1)].shape)
per_women = round(len(data[(data.gender == 0) & (data.is_promoted == 1)]) / len(data[data.gender == 0]) * 100, 2)
print(f'Percent of promoted men: {per_men}%')
print(f'Percent of promoted women: {per_women}%')
```

```python
print(data[data.awards_won > 0].shape)
print(data[(data.awards_won > 0) & (data.is_promoted == 1)].shape)
per_won = round(len(data[(data.awards_won > 0) & (data.is_promoted == 1)]) / len(data[data.awards_won > 0]) * 100, 2)
print(data[data.awards_won == 0].shape) 
print(data[(data.awards_won == 0) & (data.is_promoted == 1)].shape)
per_n_won = round(len(data[(data.awards_won == 0) & (data.is_promoted == 1)]) / len(data[data.awards_won == 0]) * 100, 2)
print(f'Percent of promoted with awards: {per_won}%')
print(f'Percent of promoted with no awards: {per_n_won}%')
```
