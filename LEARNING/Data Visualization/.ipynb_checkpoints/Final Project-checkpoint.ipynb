```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file_path = 'marvel-wikia-data.csv'

marvel_data = pd.read_csv(file_path)

marvel_data.head()
```

```python
sns.histplot(data=dc_data, x='YEAR', y='SEX')
```

```python
sns.histplot(data=marvel_data, x='Year', y='SEX')
```

```python
sns.lineplot(data=dc_data, x='YEAR', y='APPEARANCES')
```

```python
sns.lmplot(data=dc_data, x='YEAR', y='APPEARANCES')
```

```python
sns.regplot(data=dc_data, x='YEAR', y='APPEARANCES')
```

```python
file_path = 'ign_scores.csv'

ign_data = pd.read_csv(file_path, index_col='Platform')

ign_data.head()
```

```python
plt.figure(figsize=(15, 6))
sns.swarmplot(data=ign_data)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dc_data, x='YEAR', y='ALIGN')
```

```python
sns.heatmap(data=ign_data, annot=True)
```

```python
file_path = 'fifa.csv'

fifa_data = pd.read_csv(file_path, index_col='Date', parse_dates=True)

fifa_data.head()
```

```python
sns.set_style('whitegrid')
sns.barplot(data=fifa_data)
```

```python
file_path = 'museum_visitors.csv'

museum_data = pd.read_csv(file_path, index_col='Date', parse_dates=True)

museum_data.head()
```
