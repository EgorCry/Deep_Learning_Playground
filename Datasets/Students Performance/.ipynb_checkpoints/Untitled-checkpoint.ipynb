```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('exams.csv')

data.columns = data.columns.str.replace(' ', '_')

data.head()
```

```python
conditions = [(data.math_score <= 70), (data.math_score > 70)]
values = ['less than 70', 'more than 70']

data['math_score_over_70'] = np.select(conditions, values)

data.head()
```
