```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('annotations/maksssksksss0.xml')
```

```python
root = tree.getroot()
```

```python
for i in range(len(root)):
    print(i)
    for j in range(len(root[i])):
        print(f'{root[i][j].tag}: {root[i][j].text}')
        if root[i][j].tag == 'bndbox':
            for k in range(4):
                print(f'\t{root[i][j][k].tag}: {root[i][j][k].text}')
```
